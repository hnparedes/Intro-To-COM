# Question 1: Create a realistic dataset for an house market in ***Florida***.










**Part 1: Creating the data**

*   Import Numpy, Pandas, and Matplotlib
*   Set random seed to 42 using np.random.seed(42)



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)
```

Create three numpy lists called ***house_prices***, ***square_footages***, and ***locations***:


*   Create a numpy array with 300 samples (number of houses) from 100000 to 800000 for house prices using numpy random module.
*   Create a numpy array with 300 samples from 1000 to 5000 for square footages using numpy random module.
*   Create a numpy array with 300 samples from ***four*** areas which are 'Miami', 'Orlando', 'Tampa', and 'Jacksonville' using numpy random module.
*   Create a dataframe from pandas using the arrays.





```python
house = np.random.randint(100000, 800000, size=300)
square = np.random.randint(1000, 5000, size=300)

locations = np.random.choice(["Miami", "Orlando", "Tampa", "Jacksonville"], size=300)
data = {'House_Price': house, 'Square_Footage': square, 'Location': locations}
df = pd.DataFrame(data)
```

**Part 2: Analyze your dataset with the Scatter, Boxplot, Histogram, and Pie plots using Matplotlib.**

Create a ***scatter*** plot to show relationship between house prices and square footages.

*   ***xlabel*** is 'Square Footage'
*   ***ylabel*** is 'House Price ($)'
*   ***title*** is 'House Price vs Square Footage'
*   ***Color*** is 'blue'
*   ***alpha*** is 0.6
*   ***grid*** is True





```python
plt.scatter(df['Square_Footage'], df['House_Price'], color='blue', alpha=0.6)
plt.xlabel('Square Footage')
plt.ylabel('House Price ($)')
plt.title('House Price vs Square Footage')
plt.grid(True)
plt.show()
```

Create a ***histogram*** of house prices.

*   ***xlabel*** is 'House Price ($)'
*   ***ylabel*** is 'Number of Houses'
*   ***title*** is 'Distribution of House Prices'
*   ***Color*** is 'green'
*   ***edgecolor*** is 'black'
*   ***bins*** is 20
*   ***alpha*** is 0.5
*   ***grid*** is True

```python
plt.hist(df['House_Price'], bins=20, color='green', edgecolor='black', alpha=0.5)
plt.xlabel('House Price ($)')
plt.ylabel('Number of Houses')
plt.title('Distribution of House Prices')
plt.grid(True)
plt.show()
```

Create a ***boxplot*** to show house price distribution by different locations.

*   ***xlabel*** is 'Location'
*   ***ylabel*** is 'House Price ($)'
*   ***title*** is 'House Price Distribution by Different Locations'

```python
plt.figure(figsize=(6, 5))
plt.boxplot([df['House_Price'][df['Location'] == 'Miami'],
             df['House_Price'][df['Location'] == 'Orlando'],
             df['House_Price'][df['Location'] == 'Tampa'],
             df['House_Price'][df['Location'] == 'Jacksonville']],
            labels=['Miami', 'Orlando', 'Tampa', 'Jacksonville'])

plt.xlabel('Location')
plt.ylabel('House Price ($)')
plt.title('House Price Distribution by Different Locations')
plt.show()
```

Create a ***Pie Chart*** to show distribution of house locations.

*   ***title*** is 'Distribution of House Locations'

```python
location_counts = df['Location'].value_counts()
plt.figure(figsize=(5, 4))
plt.pie(location_counts, labels=location_counts.index, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightcoral', 'lightgreen', 'lightsalmon'])
plt.title('Distribution of House Locations')
plt.axis('equal')
plt.show()
```

# Question 2: Create a realistic dataset for an ***ice cream*** business.



*   Create a list of **months** from *January* to *June* (six months)
*   Create a list of flavors with 'Vanilla, 'Chocolate', 'Strawberry', and 'Mint'
*   Create a numpy array from 1000 to 5000 $ sales for each flavor and month using numpy random module.
*   Create a dataframe from pandas using the sales (**columns** are flavors and **index** is months).


```python
months = ['January', 'February', 'March', 'April', 'May', 'June']
flavors = ['Vanilla', 'Chocolate', 'Strawberry', 'Mint']

sales = np.random.randint(1000, 5000, size=(6, 4))
df = pd.DataFrame(sales, index=months, columns=flavors)
print(df)
```

Use .info(), .describe(), and .head(10) to get an idea of what the dataset looks like - show outputs for all 3 of these

Create a **Stacked Bar** chart to visualize ice cream sales over time with flavors.

*   ***xlabel*** is 'Month'
*   ***ylabel*** is 'Sales'
*   ***title*** is 'Ice Cream Sales by Flavor and Time Period'
*   ***alpha*** is 0.7
*   ***grid*** is axis='y'
*   ***rotation*** is 45

```python
plt.figure(figsize=(5, 4))

for i, flavor in enumerate(flavors):
    plt.bar(months, sales[:, i], label=flavor, alpha=0.7)

plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Ice Cream Sales by Flavor and Time Period')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.legend()
plt.show()
```

Create a **Scatter** plot to show the sales of different flavors over months.

*   ***xlabel*** is 'Month'
*   ***ylabel*** is 'Sales'
*   ***title*** is 'Ice Cream Sales by Flavor and Month'
*   ***grid*** is True
*   ***rotation*** is 45

```python
plt.figure(figsize=(5, 4))

for i, flavor in enumerate(flavors):
    plt.scatter(months, sales[:, i], label=flavor, alpha=0.7)

plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Ice Cream Sales by Flavor and Month')
plt.grid(True)
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

Create a **Pie** chart for ice cream flavor distribution in **June**.

*   ***title*** is 'Ice Cream Flavor Distribution in June'
*   ***shadow*** is True
*   ***startangle*** is 140


```python
plt.figure(figsize=(5, 4))
june_sales = sales[5]

plt.pie(june_sales, labels=flavors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Ice Cream Flavor Distribution in June')
plt.axis('equal')
plt.show()
```
